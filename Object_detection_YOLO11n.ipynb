```python
# Load the model
model = YOLO("yolo11n.pt")

# Train the model
results = model.train(
    task="detect",
    mode="train",
    data="data.yaml",       # YAML file in the current directory
    epochs=10,
    imgsz=224,
    plots=True,
    device="cpu"  # or "mps" if you're using Apple Silicon with PyTorch MPS support
)
```

```python
#Validate Custom Model
# Load the trained model
model = YOLO("runs/detect/train/weights/best.pt")

# Run validation
results = model.val(data="data.yaml")
```

```python
#prediction
# Load trained model
model = YOLO("runs/detect/train/weights/best.pt")

# Run prediction on test images
results = model.predict(source="test/images", conf=0.25, save=True)
```
